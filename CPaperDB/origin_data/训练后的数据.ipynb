````python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # 读取本地的 .env file

openai.api_key  = os.getenv('OPENAI_API_KEY')

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]

def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
#     print(str(response.choices[0].message))
    return response.choices[0].message["content"]
````

## 5.1 GPT根据相同的prompt出题

````python
messages =  [  
{'role':'system', 'content':f'{system_prompt}'},    
{'role':'user', 'content':f'{user_prompt}'},   
]
````

````python
gpt_answer = get_completion_from_messages(messages,temperature = 0)
````

````python
print(gpt_answer)
````

## 5.2 GPT对题目进行打分 并给出胜负
* 参考方式 Vicuna: An open-source chatbot impressing gpt-4 with 90%* chatgpt quality.
* 没有微调的 ChatGLM vs. 微调后的 ChatGLM
* 微调后的ChatGLM vs. GPT相同的prompt输出
* 最后计算赢的比率 Win-Tie(%) 作为微调/出题的质量

````python
score_system_prompt = """
你是检查生成题目质量的助手。
你将收到 Assistant1 和 Assistant2，两个Assistant 对于所给的要求进行问题生成。
生成的问题是直接用于考试的，因此问题不可以直接给出答案，不可以将答案写在题目中。
所给的要求、Assistant1 和 Assistant2的回答将在```分割符中给出。
你对两个Assistant的生成问题质量进行评价，提供反馈：
1.两个Assistant初始分数都是 10 分。
2.首先需要评价生成的问题，是否满足考试要求：如果直接给出答案或者将答案写在题目中，扣 5 分。
3.请评价他们的回答的准确性、有用性、相关性和详细程度，加合适的分(1分-5分)。
4.根据是否满足考试要求、有用性、无害性、相关性、准确性、创意性和详细程度等因素，给出给分过程。
5.一步步计算每个助手的给分过程，给出每个助手的最终得分。

[给分过程举例]Assistant不满足考试要求，但是出题的准确性还行，最后得分为：初始10分，不满足要求扣5分，评价回答加2分，最后得分7分。
[注意]避免任何潜在的偏见，并确保两个Assistant回答的顺序不会影响您的判断。

使用下面的格式：
Request:
<生成题目的要求>
Assistant1:
<Assistant1的回答>
Assistant2:
<Assistant2的回答>

Assistant1 Score: <Assistant1分数>
Assistant2 Score: <Assistant2分数>
Explain: <给分说明>

你只需要输出:
Assistant1 Score: <Assistant1分数>
Assistant2 Score: <Assistant2分数>
Explain: <给分说明>
"""
````

### 没有微调的 ChatGLM vs. 微调后的 ChatGLM

````python
user_message = f"""
                Request:
                ```{user_prompt}```
                Assistant1:
                ```{ft_answer}```
                Assistant2: 
                ```{uft_answer}```
                """

messages =  [  
{'role':'system', 'content':f'{score_system_prompt}'},    
{'role':'user', 'content':f'{user_message}'},   
]

response = get_completion_from_messages(messages,temperature = 0)
````

````python
print(response)
````

### 微调后的ChatGLM vs. GPT相同的prompt输出

````python
user_message = f"""
                Question:
                ```{user_prompt}```
                Assistant1:
                ```{gpt_answer}```
                Assistant2: 
                ```{ft_answer}```
                """

messages =  [  
{'role':'system', 'content':f'{score_system_prompt}'},    
{'role':'user', 'content':f'{user_message}'},   
]

response = get_completion_from_messages(messages,temperature = 0)
````

````python
print(response)
````
